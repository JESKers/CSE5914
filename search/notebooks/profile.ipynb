# Car Dataset Profiling
Kangjie (Elasticsearch). Profiles the Kaggle [CooperUnion/cardataset](https://www.kaggle.com/datasets/CooperUnion/cardataset).

Put `data.csv` in `data/` (repo root), then run all cells. Findings here drive the mapping (`search/index_mapping.py`) and cleaning rules (`search/clean_data.py`).

```python
import pandas as pd
from pathlib import Path

DATA = Path('..') / '..' / 'data' / 'data.csv'
df = pd.read_csv(DATA)
print(f'rows: {len(df):,}   columns: {df.shape[1]}')
df.head()
```

## Schema & dtypes

```python
df.info()
```

## Null counts per column

```python
nulls = df.isna().sum()
pct = (nulls / len(df) * 100).round(2)
pd.DataFrame({'nulls': nulls, 'pct': pct}).query('nulls > 0').sort_values('nulls', ascending=False)
```

## Distinct values: Make / Transmission Type / Vehicle Style

```python
for col in ['Make', 'Transmission Type', 'Vehicle Style']:
    print(f'\n=== {col}: {df[col].nunique()} distinct ===')
    print(df[col].value_counts().head(20))
```
